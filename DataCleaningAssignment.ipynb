```python
import numpy as np
import pandas as pd
import re
import nltk
import string
import spacy
from nltk.corpus import stopwords
from emot.emo_unicode import UNICODE_EMO, EMOTICONS
from bs4 import BeautifulSoup

# Reading the dataset
df = pd.read_csv(r'C:\Users\Ebbie\Downloads\TweetRaw.csv')

#lowercase the data
df = df['Messages'].str.lower()

#removing punctuation
data = df.str.replace('[^\w\s]','')

#remove stopwords
STOPWORDS = set(stopwords.words('english'))
def stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
sentence = data.apply(stopwords)

# removing emoticons
def emoticons(text):
    emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')
    return emoticon_pattern.sub(r'', text)

current = sentence.apply(emoticons)

#removing html tags
def html(text):
    return BeautifulSoup(text, "lxml").text
info = current.apply(html)

#replace non-ASCII characters with single space
def remove_non_ascii_1(text):
    return ''.join([i if ord(i) < 128 else ' ' for i in text])
semiFinal = info.apply(remove_non_ascii_1)
semiFinal
```
